```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#simulerer n tilfeldige serier av normalfordelte tall
T = 40
n = 1_000_000
x = 0.0 + .15*np.random.randn(T,n) #x iid~N(0.0, .15^2)
```

```python
#snitt og std.avvik av utvalg
print(f"{x.mean()=:.3f} and {x.std()=:.3f}")
```

```python
#tar f√∏rstedifferansen for seriekorrelert avkastning
r = np.diff(x, axis = 0)
```

```python
#snitt og std.avvik av seriekorrelert avkastning
print(f"{r.mean()=:.3f} and {r.std()=:.3f}")
```

```python
#trekker fra halvparten av variansen
r = r - 0.5*r.var()
```

```python
#forventet sluttverdier av fondet
end_values = np.exp(r.sum(axis= 0))
print(f"{end_values.mean()=:.3f} and {end_values.std()=:.3f}")
```

```python
#illustrasjon av fondsutvikling for 5 tilfeldige serier
plt.plot(np.exp(r[:,:5].cumsum(axis=0)))
plt.show()
```

```python
#sammenlikning av sluttverdier om r ikke var seriekorrelert
r = -0.5*0.21**2 + 0.21*np.random.randn(T,n) #r iid~N(-0.5*0.21^2, 0.21^2)

end_values = np.exp(r.sum(axis= 0))
print(f"{end_values.mean()=:.3f} and {end_values.std()=:.3f}")
```

```python
#illustrasjon av fondsutvikling for 5 tilfeldige serier
plt.plot(np.exp(r[:,:5].cumsum(axis=0)))
plt.show()
```
